```python
# @sync_to_async
async def createProduct(product_json):
    product = Product()
    product.asin = product_json['asin']
    product.sid = product_json.get('sid')
    product.title = product_json['title']
    product.description = product_json.get('description')
    product.item_available = product_json.get('item_available')
    product.main_image = product_json.get('main_image')
    product.save()
    print('\t'+product.asin)
    if product_json.get('categories'):
        for categ in product_json['categories']:
            category, created = Category.objects.get_or_create(category = categ['category'], url=categ['url'])
            print(category.url)
            if created:
                category.save()
            print(category.url)
            product.categories.add(category)
    
    if product_json.get('reviews'):
        rew = product_json.get('reviews')
        review, created = ProductReviewMetaData.objects.get_or_create(total_reviews = rew['total_reviews'], rating=rew['rating'], answered_questions = rew.get('answered_questions'))
        review.save()
        product.reviews =review
    
    
    if product_json.get('variants'):
        for var in product_json.get('variants'):
            if var['price'] == '':
                var['price'] = 25
            if isinstance(var.get('images'), dict):
                variant, created = ProductVariant.objects.get_or_create(asin = var['asin'], title = var['title'], image = var.get('images').get("hiRes"), price = var['price'])
            else:
                variant, created = ProductVariant.objects.get_or_create(asin = var['asin'], title = var['title'], price = float(var['price']))
            variant.save()
            product.variants.add(variant)
            
    if product_json.get('images'):
        for img in product_json['images']:
            productImage, created = ProductImage.objects.get_or_create(image=img)
            productImage.save()
            product.images.add(productImage)
    
    if product_json.get('image'):
        for img in product_json['image']:
            productImage, created = ProductImage.objects.get_or_create(image=img)
            productImage.save()
            product.images.add(productImage)

    if product_json.get('feature_bullets'):
        for bullet in product_json['feature_bullets']:
            productFeatureBullet, created = ProductFeatureBullet.objects.get_or_create(feature_bullet=bullet)
            productFeatureBullet.save()
            product.feature_bullets.add(productFeatureBullet)

            
    if product_json.get('also_bought'):
        for relatedProduct in product_json['also_bought']:
            alsoBoughtProduct = await createProduct(relatedProduct)
            alsoBoughtProduct.save()
            product.also_bought.add(alsoBoughtProduct)

    productPrice, created = ProductPrice.objects.get_or_create(**product_json['price'])
    product.price = productPrice
    print(product)
    product.save()
    return product
```

```python
# @sync_to_async
createdReview = 0
async def createProductReview(product_review_json):
    review = ProductReview()
    if Product.objects.filter(asin = product_review_json['asin']).count() > 0:
        
        review.asin = Product.objects.get(asin = product_review_json.get('asin'))
        review.rating = product_review_json.get('overall')
        review.verified = product_review_json.get('verified')
        review.reviewTime = product_review_json.get('reviewTime')
        review.reviewerID = product_review_json.get('reviewerID')
        review.reviewerName = product_review_json.get('reviewerName')
        review.reviewText = product_review_json.get('reviewText')
        review.summary = product_review_json.get('summary')

        review.save()
        print(review)

        return review
```
